```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1 = 0.02014
a2 = 0.02014 * 50
t1 = 0.9414611584602965
mj = 9.5e-4
(a1/a2)**3**0.5
```

```python
sim = rebound.Simulation()
sim.add(m=1.22)
sim.add(m=10.2*mj, a=a1,e=0)
sim.add(m=mj, a=a2)
sim.move_to_com()
```

```python
N=200
transittimes = np.zeros(N)
p = sim.particles
i = 0
while i<N:
    y_old = p[1].y - p[0].y  # (Thanks to David Martin for pointing out a bug in this line!)
    t_old = sim.t
    sim.integrate(sim.t+5e-5) # check for transits every 0.5 time units. Note that 0.5 is shorter than one orbit
    t_new = sim.t
    if y_old*(p[1].y-p[0].y)<0. and p[1].x-p[0].x>0.:   # sign changed (y_old*y<0), planet in front of star (x>0)
        while t_new-t_old>1e-11:   # bisect until prec of 1e-5 reached
            if y_old*(p[1].y-p[0].y)<0.:
                t_new = sim.t
            else:
                t_old = sim.t
            sim.integrate( (t_new+t_old)/2.)
        transittimes[i] = sim.t
        i += 1
        sim.integrate(sim.t+ 5e-2)       # integrate 0.05 to be past the transit 
```

```python
A = np.vstack([np.ones(N), range(N)]).T
c, m = np.linalg.lstsq(A, transittimes, rcond=None)[0]
```

```python
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.set_xlim([0,N])
ax.set_xlabel("Transit number")
ax.set_ylabel("TTV [s]")
plt.scatter(range(N), (transittimes-m*np.array(range(N))-c)*(3600*24.*365./2./np.pi));
```

```python
from uncertainties import ufloat

def getn(unfloat):
    return unfloat.n

def gets(unfloat):
    return unfloat.s


getn_v = np.vectorize(getn)
gets_v = np.vectorize(gets)
```

```python
tcs = getn_v(wasp18b.tcs)
epochs = wasp18b.epochs
```

```python
from scipy import optimize as op

def f_1(x, A, B):
    return A * x + B

A, B = op.curve_fit(f_1, epochs-epochs[0], tcs)[0]
```
