```python
from fitransit.base import fitlpf
import warnings
warnings.filterwarnings('ignore')
```

# 1. Set model

```python
wasp44b = fitlpf("WASP-44 b")
```

# 2. Get parameter

```python
wasp44b.get_parameter()
```

# 3. Download data

```python
wasp44b.download_data()
```

# 4. Differential evolution

```python
wasp44b.de(datadir = './data/WASP-44b/mastDownload/TESS/tess2018263035959-s0003-0000000012862099-0123-s/')
```

# 5. Single fit

```python
wasp44b.fit_singles()
```

```python
for i in range(len(wasp44b.lpf.times)):
    fig = wasp44b.singles[i].plot_single_transit(ylim=(0.97,1.02))
```
