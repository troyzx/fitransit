```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse,Arc
```

```python
time = np.linspace(0,100,1001)
```

```python
time
```

```python
y = lambda x:(1. if x <30 else 1.24-0.008*x if x<35 else 0.96 if x<60 else 0.48+0.008*x if x<65 else 1)
flux = np.vectorize(y)
```

```python
fig,ax = plt.subplots(figsize=(6.5,2.5),dpi=200)
plt.plot(time,flux(time))

plt.axvline(30,ymax=0.1/0.22,c='k',ls='--')
plt.axvline(35,ymax=0.06/0.22,c='k',ls='--')
plt.axvline(60,ymax=0.06/0.22,c='k',ls='--')
plt.axvline(65,ymax=0.1/0.22,c='k',ls='--')
plt.text(2.7,1.09,"star")
# plt.text(-4.5,1.02,"planet")


star1 = Ellipse((5, 1.06),8,0.041,color='r')
ax.add_patch(star1)

planet1 = Ellipse((-3, 1.06),1.2,0.006,color='k')
ax.add_patch(planet1)

star2 = Ellipse((30, 1.06),8,0.041,color='r')
ax.add_patch(star2)

planet2 = Ellipse((25.3, 1.06),1.2,0.006,color='k')
ax.add_patch(planet2)

star3 = Ellipse((47.5, 1.06),8,0.041,color='r')
ax.add_patch(star3)

planet3 = Ellipse((47.5, 1.06),1.2,0.006,color='k')
ax.add_patch(planet3)

star4 = Ellipse((65, 1.06),8,0.041,color='r')
ax.add_patch(star4)

planet4 = Ellipse((69.7, 1.06),1.2,0.006,color='k')
ax.add_patch(planet4)

star5 = Ellipse((90, 1.06),8,0.041,color='r')
ax.add_patch(star5)

planet5 = Ellipse((98, 1.06),1.2,0.006,color='k')
ax.add_patch(planet5)

plt.annotate('planet', xy=(-3, 1.06), xytext=(2, 1.015),arrowprops=dict(width=1,headwidth=5,headlength=5,facecolor='black', shrink=0.2))

plt.xlabel("time")
plt.ylabel("flux                       ")

plt.ylim(0.9,1.12)
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())

plt.savefig("image/transit.pdf")
```
